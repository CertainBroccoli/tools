```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

def fourier_phase_scramble(img_array):
    """Core algorithm: Preserve amplitude energy, scramble phase."""
    # 1. 2D Fast Fourier Transform (FFT)
    fft_img = np.fft.fft2(img_array)
    
    # 2. Extract magnitude spectrum (preserves spatial frequency energy)
    magnitude = np.abs(fft_img)
    
    # 3. Generate random phase
    noise = np.random.rand(*img_array.shape)
    random_phase = np.angle(np.fft.fft2(noise))
    
    # 4. Combine magnitude with random phase, and perform inverse FFT
    scrambled_fft = magnitude * np.exp(1j * random_phase)
    scrambled_img = np.real(np.fft.ifft2(scrambled_fft))
    
    # 5. Normalize pixels to 0-255 range
    img_min = np.min(scrambled_img)
    img_max = np.max(scrambled_img)
    if img_max > img_min:
        scrambled_img = (scrambled_img - img_min) / (img_max - img_min) * 255.0
        
    return scrambled_img.astype(np.uint8)

# ==========================================
#          Jupyter Interactive Area
# ==========================================

print("🌟 Fourier Phase Scrambling Tool 🌟")

# Get user input for the image path
raw_input = input("👉 Please enter the image path: ")

# Clean path: remove leading/trailing spaces and quotes added by drag-and-drop
image_path = raw_input.strip().strip('\'"')

if not image_path:
    print("❌ No path entered. Please run the cell again.")
elif not os.path.exists(image_path):
    print(f"\n❌ Error: File '{image_path}' not found.")
else:
    print(f"\n✅ Image found. Performing Fourier Transform...")
    try:
        # Read image and convert to grayscale
        img = Image.open(image_path).convert('L')
        img_array = np.array(img)
        
        # Perform scrambling
        scrambled_array = fourier_phase_scramble(img_array)
        
        # Use matplotlib to display images directly below the Jupyter cell
        plt.figure(figsize=(12, 6))
        
        plt.subplot(1, 2, 1)
        plt.title("Original Image", fontsize=14)
        plt.imshow(img_array, cmap='gray')
        plt.axis('off')
        
        plt.subplot(1, 2, 2)
        plt.title("Phase Scrambled", fontsize=14)
        plt.imshow(scrambled_array, cmap='gray')
        plt.axis('off')
        
        plt.tight_layout()
        plt.show()
        
        print("🎉 Processing complete!")
        
    except Exception as e:
        print(f"\n❌ Error processing the image: {e}")
```
